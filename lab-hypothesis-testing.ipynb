**Challenge 1**

In this challenge, we will be working with pokemon data. The data can be found here:

- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/pokemon.csv

```python
#libraries
import pandas as pd
import scipy.stats as st
import numpy as np

```

```python
df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/pokemon.csv")
df
```

- We posit that Pokemons of type Dragon have, on average, more HP stats than Grass. Choose the propper test and, with 5% significance, comment your findings.

```python
# H0: Dragon avg HP = Grass avg HP
# H1: Dragon avg HP > Grass avg HP

# Nivel de significancia 0.05

dragon = (df['Type 1'] == "Dragon") | (df['Type 2'] == "Dragon")
grass = (df['Type 1'] == "Grass") | (df['Type 2'] == "Grass")

# 2. Aplicamos las condiciones para obtener las series de HP
Dragon_avg_stats = df[dragon]["HP"]
Grass_avg_stats = df[grass]["HP"]
```

```python
_, p_value = st.ttest_ind(Dragon_avg_stats, Grass_avg_stats, equal_var=False)
print(p_value)
```

### resultado
Con este resultado de p_value, rechazamos la hipótesis nula y aceptamos la hipótesis alternativa, es decir que hay una diferencia estadística significativa para concluir que la HP media de los pokemon tipo Dragon es mayor que la de los pokemon tipo Grass

- We posit that Legendary Pokemons have different stats (HP, Attack, Defense, Sp.Atk, Sp.Def, Speed) when comparing with Non-Legendary. Choose the propper test and, with 5% significance, comment your findings.


```python
# H0 stats_legendary = stats_non_legendary
# H1 stats_legendary != stats_non_legendary

# Nivel de significancia 0.05

legendary_pokemon = df[df['Legendary'] == True]
non_legendary_pokemon = df[df['Legendary'] == False]

stats_column = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']

for stat in stats_column:   
    legendary_data = legendary_pokemon[stat]
    non_legendary_data = non_legendary_pokemon[stat]

   
    _, p_value = st.ttest_ind(legendary_data, non_legendary_data, equal_var=True)

p_value
```

### Resultado
Rechazamos la hipótesis nula y aceptamos la hipótesis alternativa. Se comprueba estadísticamente que los pokemon Legendary tienen un nivel de stats superior al de los pokemon Non-Legendery

In this challenge, we will be working with california-housing data. The data can be found here:

```python
df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/california_housing.csv")
df.head()
```

**We posit that houses close to either a school or a hospital are more expensive.**

- School coordinates (-118, 34)
- Hospital coordinates (-122, 37)

We consider a house (neighborhood) to be close to a school or hospital if the distance is lower than 0.50.

Hint:
- Write a function to calculate euclidean distance from each house (neighborhood) to the school and to the hospital.
- Divide your dataset into houses close and far from either a hospital or school.
- Choose the propper test and, with 5% significance, comment your findings.
 

```python
def dist_euclidiana(df, coord_sch_hosp):
 
    sch_hosp_lon, sch_hosp_lat = coord_sch_hosp 
    
    dif_lon = df['longitude'] - sch_hosp_lon
    dif_lat = df['latitude'] - sch_hosp_lat
    
    # Aplica el Teorema de Pitágoras: sqrt((x2-x1)^2 + (y2-y1)^2)
    distance = np.sqrt(dif_lon**2 + dif_lat**2)
    
    return distance
```

```python
sch_coord = (-118, 34)
hosp_coord = (-122, 37)

df['dist_to_school'] = dist_euclidiana(df, sch_coord)
df['dist_to_hospital'] = dist_euclidiana(df, hosp_coord)

distancia_ref = 0.50

df['cercania'] = (df['dist_to_school'] <= distancia_ref) | (df['dist_to_hospital'] <= distancia_ref)

casas_cerca = df[df['cercania'] == True]['median_house_value']
casas_lejos = df[df['cercania'] == False]['median_house_value']
```

```python
# H0 Valor medio de las casas cercanas = Valor medio de las casas lejanas
# H1 Valor medio de las casas cercanas > Valor medio de las casas lejanas

# Nivel de significancia 0.05

_, p_value = st.ttest_ind(casas_cerca, casas_lejos,equal_var=False)

print(p_value)
```

### Resultado

Al hacer todo el análisis podemos ver que el p_value es muchísimo menor que el valor de significancia, por lo cual rechazamos la hipótesis nula y nos quedamos con la alternativa.
Las casas cercanas a un colegio ó a un hospital tienen un valor medio significativamente más alto que las casas que se encuentran más lejos de esos puntos de interés.
